
Lets change Price to Numeric


```python
import re

def pkr_to_int(s):
    if pd.isna(s):
        return None

    # Make sure it's a string and remove newlines + "PKR"
    s = str(s).replace('\n', ' ').replace('PKR', '').strip()

    # Extract the numeric part
    m = re.search(r'([\d.]+)', s)
    if not m:
        return None

    value = float(m.group(1))
    text = s.lower()
    # Decide multiplier based on unit
    if 'crore' in text:
        mult = 10_000_000     # 1 crore = 10,000,000
    elif 'thousand' in text:
        mult = 1_000          # 1 thousand = 1,000
    elif 'lakh' in text:
        mult = 100000
    elif 'arab' in text:
      mult=1000000000
    else:
      mult=1
              # plain number, no unit

    return int(round(value * mult))


```

Lets bring Area into Standard Scale

```python

def area_to_sqyd(s):
    """
    Convert area strings like:
      - '5 Marla'
      - '1 Kanal'
      - '1 Kanal 5 Marla'
      - '2.5 Kanal'
    into square yards (float).

    Assumes:
      1 Marla = 30.25 sq. yd
      1 Kanal = 605 sq. yd  (20 Marla)
    """
    if pd.isna(s):
        return None

    # Normalize text: remove newlines, strip spaces, lowercase
    text = str(s).replace('\n', ' ').strip().lower()
    m_sqyd = re.search(
    r'([\d.]+)\s*(sq\.?\s*yds?|sq\.?\s*yd|sqyards?|sq\s*yards?)',
    text)
    if m_sqyd:
    # Just return the numeric part as float; already in sq. yd
      return float(m_sqyd.group(1))

    # Find all "<number> <unit>" pairs, e.g. "1 kanal", "5 marla"
    matches = re.findall(r'([\d.]+)\s*(kanal|kanals|marla|marlas)', text)
    if not matches:
        # No recognizable unit; you can return None or try to interpret as plain sq. yd
        return None

    total_sqyd = 0.0

    for value_str, unit in matches:
        value = float(value_str)

        if unit.startswith('kanal'):
            total_sqyd += value * 605       # 1 kanal = 605 sq. yd
        elif unit.startswith('marla'):
            total_sqyd += value * 30.25     # 1 marla = 30.25 sq. yd

    return total_sqyd
```

```python
excel_df['Area_sqyd'] = excel_df['Area'].apply(area_to_sqyd)
```

```python
# Outlier Detection (IQR for 'Age' and 'Fare')
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

price_outliers = detect_outliers_iqr(excel_df['Price_Amount'])
area_outliers = detect_outliers_iqr(excel_df['Area_sqyd'])
```

```python
def outlier_mask_iqr(series, factor=1.5):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - factor * IQR
    high = Q3 + factor * IQR
    mask = (series < low) | (series > high)
    return mask, low, high

price_mask, _, _ = outlier_mask_iqr(excel_df['Price_Amount'])
area_mask,  _, _ = outlier_mask_iqr(excel_df['Area_sqyd'])

combined_mask = price_mask | area_mask          # True = outlier in price or area
excel_df_clean = excel_df[~combined_mask].copy()
```

```python
df = excel_df_clean

size_col   = 'Area_sqyd'
price_col  = 'Price_Amount'
city_col   = 'City'
ptype_col  = 'Type'


cols_needed = [size_col, price_col, city_col, ptype_col]
data = df[cols_needed].dropna()

data = data[(data[size_col] > 0) & (data[price_col] > 0)]


def size_price_corr(group):

    if group[size_col].nunique() < 2 or group[price_col].nunique() < 2:
        return np.nan
    return group[size_col].corr(group[price_col])  # Pearson by default

corr_df = (
    data
    .groupby([city_col, ptype_col])
    .apply(size_price_corr)
    .reset_index(name='size_price_corr')
)

corr_matrix = corr_df.pivot(index=city_col, columns=ptype_col, values='size_price_corr')

corr_matrix = corr_matrix.dropna(how='all', axis=0)  # drop cities with all NaNs
corr_matrix = corr_matrix.dropna(how='all', axis=1)  # drop property types with all NaNs


fig, ax = plt.subplots(figsize=(10, 6))

im = ax.imshow(corr_matrix.values, aspect='auto', vmin=-1, vmax=1)

ax.set_xticks(range(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns, rotation=45, ha='right')

ax.set_yticks(range(len(corr_matrix.index)))
ax.set_yticklabels(corr_matrix.index)

ax.set_title('Correlation between Property Size and Price\nby City and Property Type')
ax.set_xlabel('Property Type')
ax.set_ylabel('City')

cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Pearson correlation (size vs price)')

plt.tight_layout()
plt.show()
```
